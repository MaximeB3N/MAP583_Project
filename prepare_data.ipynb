```python
import os
import sys
from pathlib import Path
import argparse
import pandas as pd
from tqdm import tqdm
```

```python
root_dir = "/Volumes/Mémoire externe Maxime/Dataset/TAU_DATASET"
output_dir = "/Volumes/Mémoire externe Maxime/Dataset/TAU_SMALL_DATASET/data"
file_list = "meta_data/fold1_train.csv"
list_train = "meta_data/fold1_train.csv"
list_test = "meta_data/fold1_test.csv"
```

```python
if not os.path.exists(root_dir):
        print("Root directory doesn't exist")
        sys.exit(1)

if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
pd.read_csv(file_list, sep="	")["filename"].to_list()
```

```python
# create a list of all the files in the root directory
root_list = [Path(root_dir) / folder for folder in os.listdir(root_dir) 
                if os.path.isdir(Path(root_dir) / folder) and os.path.exists(Path(root_dir) / f"{folder}/audio")]
paths_list = [ Path(folder) / f"audio/{name}" for folder in root_list for name in os.listdir(folder/"audio") if name.endswith(".wav")]


# read the file_list_evaluate file
to_be_kept = [ path for path in paths_list if "audio/"+path.name in pd.read_csv(file_list, sep="	")["filename"].to_list()]
```

```python
len(pd.read_csv(file_list, sep="	")["filename"].to_list())
```

```python
len(to_be_kept)
```

```python
print('Moving files to output directory')
for path in tqdm(to_be_kept):
    os.rename(path, Path(output_dir) / path.name)
print("Done")
```

```python
# loading it and remove 'audio' from the filename
list_test = pd.read_csv(list_test, sep="	").apply(lambda x: x["filename"].replace("audio/", ""), axis=1).to_list()
```

```python
if not os.path.exists(Path(output_dir).parent / "train"):
    os.mkdir(Path(output_dir).parent / "train")

# the same with the test directory
if not os.path.exists(Path(output_dir).parent / "test"):
    os.mkdir(Path(output_dir).parent / "test")
```

```python
for path_train in tqdm(list_train):
    os.rename(Path(output_dir) / path_train, Path(output_dir).parent / "train" / path_train)

for path_test in tqdm(list_test):
    os.rename(Path(output_dir) / path_test, Path(output_dir).parent / "test" / path_test)
```
